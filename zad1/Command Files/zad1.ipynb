```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#wczytanie bazy danych
df=pd.read_csv("/home/michal/AiBD/lab5/TIER/Original Data/drinks.csv")
```

```python
#wybranie poszczególnych kolumn, posortowanie i wybranie 20 najlepszych wyników
q1=df[['country','total_litres_of_pure_alcohol']]
q1=q1.sort_values(ascending=False,by=['total_litres_of_pure_alcohol'])
q1=q1.iloc[:20]
```

```python
#wyświetlenie wyników na wykresie
x = np.arange(len(q1.values[:,0]))
width = 0.5

fig, ax = plt.subplots()
rects1 = ax.bar(x, q1.values[:,1], width)

ax.set_ylabel('Amount')
ax.set_title('Amount of total alcohol drunk in top 20 countries')
ax.set_xticks(x)
ax.set_xticklabels(q1.values[:,0])

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)

fig.tight_layout()
plt.rcParams['figure.figsize'] = [20, 10]
plt.show()
```
