```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import mpld3

#wczytanie danych z pliku
df=pd.read_csv("/home/michal/AiBD/lab5/earthquake_data.csv")

#wybranie interesujących nas kolumn
q1=df[['In general, how worried are you about earthquakes?','US Region']]
str1 = 'In general, how worried are you about earthquakes?'
#wybranie pozytywnych odpowiedzi na pytanie
q1=q1.loc[(q1[str1] == 'Extremely worried') | (q1[str1] == 'Very worried') | (q1[str1] == 'Somewhat worried')]
#pogrupowanie dancych
q1=q1.groupby(['US Region']).count()
A = np.squeeze(np.asarray(q1.values))
```

```python
#wyświetlenie wyników
x = np.arange(len(q1.index))
width = 0.5

fig, ax = plt.subplots()
rects1 = ax.bar(x, A, width, label='People afraid')

ax.set_ylabel('Amount')
ax.set_title('In general, how worried are you about earthquakes?')
ax.set_xticks(x)
ax.set_xticklabels(q1.index)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)

fig.tight_layout()
plt.rcParams['figure.figsize'] = [15, 10]
plt.show()
```
